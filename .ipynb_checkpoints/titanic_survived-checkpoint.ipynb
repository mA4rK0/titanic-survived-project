## Module

```python
class TitanicFeatureEngineer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.age_median_ = None
        self.ticket_count_ = None
        self.global_age_median_ = None

    def fit(self, X, y=None):
        X = X.copy()

        self.global_age_median_ = X["Age"].median()

        self.age_median_ = (
            X.groupby(["Sex", "Pclass"])["Age"]
             .median()
        )

        self.ticket_count_ = X["Ticket"].value_counts()

        return self

    def transform(self, X):
        X = X.copy()

        X["Title"] = X["Name"].str.extract(r" ([A-Za-z]+)\.", expand=False)
        rare_titles = ["Lady","Countess","Capt","Col","Don","Dr","Major","Rev","Sir","Jonkheer","Dona"]
        X["Title"] = X["Title"].replace(rare_titles, "Rare")
        X["Title"] = X["Title"].replace({"Mlle":"Miss","Ms":"Miss","Mme":"Mrs"})
        X["Title"] = X["Title"].fillna("Unknown")

        X["FamilySize"] = X["SibSp"].fillna(0) + X["Parch"].fillna(0) + 1
        X["IsAlone"] = (X["FamilySize"] == 1).astype(int)

        X["FarePerPerson"] = X["Fare"] / X["FamilySize"]
        X["FareLog"] = np.log1p(X["Fare"])

        X["CabinDeck"] = X["Cabin"].astype(str).str[0]
        X["CabinDeck"] = X["CabinDeck"].replace("n", np.nan)
        X["HasCabin"] = (X["Cabin"].notna()).astype(int)

        X["TicketGroupSize"] = X["Ticket"].map(self.ticket_count_).fillna(1).astype(int)

        def fill_age(row):
            if pd.notna(row["Age"]):
                return row["Age"]
            key = (row["Sex"], row["Pclass"])
            if key in self.age_median_.index and pd.notna(self.age_median_.loc[key]):
                return self.age_median_.loc[key]
            return self.global_age_median_

        X["Age"] = X.apply(fill_age, axis=1)

        return X
```

## Machine Learning Training

```python
train_df = pd.read_csv("train.csv")
test_df  = pd.read_csv("test.csv")

y = train_df[target].astype(int)
X = train_df.drop(columns=[target])
```

```python
RANDOM_STATE = 42

num_cols = ["Age","SibSp","Parch","Fare","FareLog","FamilySize","FarePerPerson","TicketGroupSize", "HasCabin"]
cat_cols = ["Pclass","Sex","Embarked","Title","IsAlone","CabinDeck"]

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ],
    remainder="drop"
)
```

```python
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=RANDOM_STATE)
scoring = {"acc": "accuracy"}

models = {
    "LogReg": LogisticRegression(max_iter=5000, random_state=RANDOM_STATE),
    "SVM_RBF": SVC(kernel="rbf", probability=False, random_state=RANDOM_STATE),
    "RF": RandomForestClassifier(n_estimators=800, random_state=RANDOM_STATE, n_jobs=-1),
    "GBC": GradientBoostingClassifier(random_state=RANDOM_STATE),
    "HGB": HistGradientBoostingClassifier(random_state=RANDOM_STATE),
    "XGB": XGBClassifier(
        random_state=RANDOM_STATE,
        n_estimators=1200,
        max_depth=4,
        learning_rate=0.03,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_lambda=1.0,
        reg_alpha=0.0,
        min_child_weight=1,
        gamma=0,
        eval_metric="logloss"
    ),
}

results = []
for name, model in models.items():
    pipe = Pipeline(steps=[
        ("fe", TitanicFeatureEngineer()),
        ("prep", preprocess),
        ("model", model)
    ])
    out = cross_validate(pipe, X, y, cv=cv, scoring=scoring, n_jobs=-1)
    results.append({
        "model": name,
        "acc_mean": out["test_acc"].mean(),
        "acc_std": out["test_acc"].std()
    })

results_df = pd.DataFrame(results).sort_values("acc_mean", ascending=False)
results_df
```

```python
ax = results_df.sort_values("acc_mean").plot(kind="barh", x="model", y="acc_mean", legend=False, figsize=(8,5))
ax.set_xlabel("CV Accuracy (mean)")
```

```python
best_name = results_df.iloc[0]["model"]
best_model = models[best_name]

base_pipe = Pipeline(steps=[
    ("fe", TitanicFeatureEngineer()),
    ("prep", preprocess),
    ("model", best_model)
])

param_dist = {}

if best_name == "XGB":
    param_dist = {
        "model__n_estimators": [700, 1000, 1300, 1600, 2000],
        "model__max_depth": [2, 3, 4, 5],
        "model__learning_rate": [0.01, 0.02, 0.03, 0.05],
        "model__subsample": [0.7, 0.8, 0.9, 1.0],
        "model__colsample_bytree": [0.6, 0.7, 0.8, 0.9, 1.0],
        "model__reg_lambda": [0.5, 1.0, 2.0, 5.0],
        "model__reg_alpha": [0.0, 0.01, 0.1],
        "model__min_child_weight": [1, 2, 5, 10],
        "model__gamma": [0, 0.1, 0.2, 0.5],
    }
elif best_name == "RF":
    param_dist = {
        "model__n_estimators": [500, 800, 1200],
        "model__max_depth": [None, 4, 6, 8, 10],
        "model__min_samples_split": [2, 5, 10],
        "model__min_samples_leaf": [1, 2, 4],
        "model__max_features": ["sqrt", "log2", None]
    }
elif best_name == "LogReg":
    param_dist = {
        "model__C": [0.1, 0.3, 1.0, 3.0, 10.0],
        "model__solver": ["lbfgs", "liblinear"]
    }
elif best_name == "SVM_RBF":
    param_dist = {
        "model__C": [0.3, 1, 3, 10, 30],
        "model__gamma": ["scale", "auto"]
    }

search = RandomizedSearchCV(
    base_pipe,
    param_distributions=param_dist,
    n_iter=min(40, max(1, len(list(param_dist.values())[0]) if param_dist else 1)),
    scoring="accuracy",
    cv=cv,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1
)

search.fit(X, y)

print("Best model:", best_name)
print("Best CV acc:", search.best_score_)
print("Best params:", search.best_params_)

best_pipe = search.best_estimator_
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

best_pipe.fit(X_train, y_train)
pred = best_pipe.predict(X_test)

print("Holdout Accuracy:", accuracy_score(y_test, pred))
print("Holdout F1:", f1_score(y_test, pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, pred))
print("\nReport:\n", classification_report(y_test, pred))

if hasattr(best_pipe.named_steps["model"], "predict_proba"):
    proba = best_pipe.predict_proba(X_test)[:, 1]
    print("Holdout ROC-AUC:", roc_auc_score(y_test, proba))

best_pipe.fit(X, y)
```
